**Question 1 : What is the difference between multithreading and multiprocessing?**

Ans :
Multithreading  allows multiple threads to run within a single process, sharing  the same memory space. It is useful for I/O bound tasks but limited by the Global Interpreter Lock (GIL) in python.
Multiprocessing creates separate processes , each with its own python interpreter and memory space, allowing true parallel execution on multiple CPU cores. It is best suited for CPU bound tasks.


**Question 2: What are the challenges associated with memory management in Python?**

Ans.
**Garbage collection overhead** : The automatic garbage collector can introduce delays.

**Reference cycles** : Circular references may prevent objects from being released promptly.

**Memory Leaks**: Poorly managed references (e.g., in global variables or circular references) can lead to memory leaks.

**Fragmentation:** Long-running applications may suffer from heap fragmentation.

**Large Objects:** Handling big data structures (like large NumPy arrays or data frames) can consume excessive memory.

**Global Interpreter Lock (GIL):** Limits true parallelism, impacting multi-threaded performance.

```python
#Q3. Write a Python program that logs an error message to a log file when a division by zero exception occurs?

import logging
logging.basicConfig(filename = 'error_log.txt', level = logging.ERROR,
                    format = '%(asctime)s - %(levelname)s - %(message)s')


try:
    a = 10
    b = 0
    result = a/b
except ZeroDivisionError as e:
        logging.error("Division by zero occurred: %s", e)
```

```python
#Q4.Write a Python program that reads from one file and writes its content to another file?
try:
    with open('source.txt', 'r') as src:
        content = src.read()
    with open('destination.txt', 'w') as dest:
        dest.write(content)
    print("File copied successfully!")
except FileNotFoundError:
    print("Source file not found.")
```

```python
#Q5. Write a program that handles both IndexError and KeyError using a try-except block?
data_list = [10, 20, 30]
data_dict = {"a": 1, "b": 2}

try:
    print(data_list[5])  # IndexError
    print(data_dict["c"])  # KeyError
except IndexError:
    print("IndexError: Tried to access an invalid list index.")
except KeyError:
    print("KeyError: The specified key does not exist in the dictionary.")
```

```python
#Q7.Explain the difference between apply() and map() in Pandas?
#map() -  is used on Series objects to apply a function element-wise.
#apply() - can be used on both DataFrames and Series, and can apply more complex operations across rows or columns.

import pandas as pd

s = pd.Series([1, 2, 3])
print(s.map(lambda x: x**2))  # map() on Series

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df.apply(lambda x: x.sum()))  # apply() on DataFrame
```

```python
#Q8.Create a histogram using Seaborn to visualize a distribution?


import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")
sns.histplot(data['sepal_length'], kde=True, color='skyblue')
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
#Q10. Calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
